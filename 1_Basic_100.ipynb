#1097

```
mat = []
for i in range(19):
    mat.append(input().split(" "))
    mat[i] = [int(mat[i][j]) for j in range(19)]

num = int(input())
for i in range(num):
    x, y = input().split(" ")
    x, y = int(x), int(y)

    for j in range(19):
        mat[x-1][j]=1 if mat[x-1][j]== 0 else 0
    for j in range(19):
        mat[j][y-1]=1 if mat[j][y-1]== 0 else 0    

for i in range(19):
    for j in range(19):
        if j == 19-1:
            print(mat[i][j])
        else:    
            print(mat[i][j], end=" ")        
```

#1098

```
h, w = input().split(" ")
h, w = int(h), int(w)

n = int(input())
pan = [[0 for i in range(w)] for j in range(h)]

for i in range(n):
    l, d, x, y = input().split(" ")
    l, d, x, y = int(l), int(d), int(x), int(y)

    xi = x-1
    yi = y-1
    if d==0:        # 가로
        pan[xi][yi:yi+l] = [1]*l
        
    else:
        for j in range(l):
            pan[xi+j][yi] = 1 

for i in range(h):
    for j in range(w):
        if j == w-1:
            print(pan[i][j])
        else:    
            print(pan[i][j], end=" ")        
```

```
miro = []
for h in range(10):
    miro.append(input().split(" "))
    miro[h] = [int(miro[h][i]) for i in range(len(miro[h]))]

y, x = 1, 1
bab = miro[y][x]
miro[y][x] = 9
while bab != 2:
    
    if miro[y][x+1] == 0:
        bab = miro[y][x+1]
        miro[y][x+1] = 9
        x += 1
        
    elif miro[y][x+1] == 2:
        miro[y][x+1] = 9
        break

    else:     
        if miro[y+1][x]==1: break
        bab = miro[y+1][x]
        miro[y+1][x] = 9
        y += 1
    
    
for i in range(10):
    for j in range(10):
        if j == 10-1:
            print(miro[i][j])
        else:    
            print(miro[i][j], end=" ")        
```
